```python
import json
import os

data = pd.read_csv('data-open-voice/annotations/data.csv', chunksize=1250)
```

```python
i = 0
for chunk in data:
    i += 1
    data_files = []
    for index, row  in chunk.iterrows():
        aud_array, sr = librosa.load(row['path'], sr=16000)
        entry = {'path':row['path'], 'text':row['sentence'], 'array':aud_array.tolist(), 'sampling_rate':sr}
        data_files.append(entry)
    
    with open(f'data-open-voice/annotations/dataset{i}.json', 'w') as f:
        json.dump(data_files, f, indent=4)
        
```
