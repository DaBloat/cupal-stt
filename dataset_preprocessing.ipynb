# Data Preprocessing

```python
df = pd.read_csv('data-open-voice/annotations/data.csv')
df.head(10)
```

```python
# Saving audio dictionary
df['audio'] = df['path'].apply(lambda x: librosa.load(x, sr=16000))
df['array'] = df['audio'].apply(lambda x: x[0])
df['sampling_rate'] = df['audio'].apply(lambda x: x[1])
df.drop(['audio'], axis=1, inplace=True)
```
